Imports

```python
import glob
# import nbconvert
import calendar
import pandas as pd
import plotly.express as px
import plotly.io as io
io.templates.default = 'plotly_dark' #Doesn't matter
from jupyter_dash import JupyterDash
import plotly.graph_objects as go
from dash import html
from dash import dcc
import dash_extensions as de
from dash.dependencies import Input, Output
```

```python
#Create enlisted df, clean it, fill it, and change pft date to datetime
enlisted_df = pd.concat(pd.read_csv(i, skiprows= 2) for  i in glob.glob(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\Assets\Enlisted_new\[0-9]*.csv'))
enlisted_df = enlisted_df.drop('1', axis = 1)
enlisted_df['MOS'].fillna('ffill', inplace= True)
enlisted_df['PFT SCORE'].fillna(enlisted_df['PFT SCORE'].mean(), inplace= True)
enlisted_df.PFT_DATE = pd.to_datetime(enlisted_df.PFT_DATE)
enlisted_df.PFT_DATE = [i.month for i in enlisted_df['PFT_DATE']]

#Create officer df, clean it, fill it, and change pft date to datetime
officer_df = pd.concat(pd.read_csv(i, skiprows= 2) for  i in glob.glob(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\Assets\Officer_new\[0-9]*.csv'))
officer_df = officer_df.drop('1', axis = 1)
officer_df['MOS'].fillna('ffill', inplace= True)
officer_df['PFT SCORE'].fillna(officer_df['PFT SCORE'].mean(), inplace= True)
officer_df.PFT_DATE = pd.to_datetime(officer_df.PFT_DATE)
officer_df.PFT_DATE = [i.month for i in officer_df['PFT_DATE']]

#Create a combined dataframe from enlisted and officer. Change PFT date to month format
combined_df = pd.concat([enlisted_df, officer_df])
```

Build the app

```python
url = "https://assets4.lottiefiles.com/packages/lf20_PviSzU.json"
# url2 = ""
options = dict(loop=True, autoplay=True, rendererSettings=dict(preserveAspectRatio='xMidYMid slice'))

app = JupyterDash(__name__) #can call whatever
app.scripts.config.serve_locally = True
app.layout = html.Div([
    html.H1('Visualizing the Physical Fitness Test', id = 'title', style={'textAlign': 'center'}),

#GIF
html.Div(de.Lottie(options=options, width="10%", height="10%", url=url)),

#Title of page 
html.Div(children=[
        html.H3('A collection of visualizations to explore different areas of the USMC PFT', id = 'title2', style={'textAlign': 'center'}),
        'This is the text to the article above']),


#Create dropdown
dcc.Dropdown(id = 'slct', options = [
    {'label': 'Enlisted', 'value': '1'},  #User will see the 'label' in the drop down. The 'value' gets returned
    {'label': 'Officer', 'value': '2'},
    {'label': 'Officer and Enlisted', 'value': '3'}],
    multi = False,
    value = '1', #This line sets the default value
    style = {'width': '35%', 'color': 'black', 'menuItemStyle': {'color':'black'}}
),
#Graph with dropdown
html.H3('Average PFT scores of Officers and Enlisted', id = 'avg', style = {'text-align': 'center', 
                                                                            'text-decoration': 'underline'}),
    html.Div([
        dcc.Graph(id = 'avg_graph', figure = {})
]),


html.Div([html.H3('Officer Vs. Enlisted Average PFT', id = 'title2', style={'textAlign': 'center'}),
    html.Iframe(srcDoc=open(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\Officer Vs. Enlisted Average PFT.html', 
    encoding = 'utf-8').read(), width ='100%' , height = 650, id = 'graph') 
]),

#POG fig
html.Div([html.H3('Who runs a better PFT? POG or non-POG', id = 'title2'),
    html.Iframe(srcDoc=open(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\Assets\Enlisted_new\POG_V_GRUNT.html', 'r', 
 encoding = 'utf-8').read(), width='49%', height=600, id = 'graph',style = {'float':'left'})
]),

html.Div(
    html.Iframe(srcDoc=open(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\PFT Classes and Failures Pie.html', 
 encoding = 'utf-8').read(), width ='49%' , height = 600, id = 'graph',style = {'float':'right'}) 
),

#MOS average PFT
html.Div([html.H3('MOS Average PFT Scores', id = 'title2',style = {'float':'left'}),
    html.Iframe(srcDoc=open(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\MOS AVERAGE PFT SCORES.html', 
    encoding = 'utf-8').read(), width ='100%' , height = 650, id = 'graph') 
]),


#Create dropdown 2
dcc.Dropdown(id = 'slct2', options = [
    {'label': 'Enlisted', 'value': '1'},  #User will see the 'label' in the drop down. The 'value' gets returned
    {'label': 'Officer', 'value': '2'},
    {'label': 'Officer and Enlisted', 'value': '3'}],
    multi = False,
    value = '1', #This line sets the default value
    style = {'width': '35%', 'color': 'black', 'menuItemStyle': {'color':'black'}}
),
#Graph with dropdown2
html.H3('Average Officer and Enlisted PFT Score in Season', id = 'avg', style = {'text-align': 'center', 
                                                                            'text-decoration': 'underline'}),
    html.Div([
        dcc.Graph(id = 'avg_graph2', figure = {})
]),

# #Custom fig
# html.Div([html.H3('What is the best month to run a PFT?', id = 'title2', style={'textAlign': 'center'}),
#     html.Iframe(srcDoc=open(r'C:\Users\six6i\OneDrive\Documents\2652\GitHub Desktop Repository\Z-Bohm\Prereq_Class\Emily Dashboard\PFT_Scores_by_Month.html', 
#     encoding = 'utf-8').read(), width ='100%' , height = 650, id = 'graph') 
# ]),

#Supplemental material

#Create dropdown 3
dcc.Dropdown(id = 'slct3', options = [
    {'label': '-Select-', 'value': '1'},  #User will see the 'label' in the drop down. The 'value' gets returned
    {'label': 'Requirements Document', 'value': '2'},
    {'label': 'Source Code', 'value': '3'}],
    multi = False,
    value = '1', #This line sets the default value
    style = {'width': '35%', 'color': 'black', 'menuItemStyle': {'color':'black'}}
),
html.Div([html.H3('Supplemental Material', id = 'title2', style={'textAlign': 'center'}),
# html.Div([html.A("Supplemental Material (Click me!), href = 'file:///C:/Users/six6i/OneDrive/Documents/2652/GitHub%20Desktop%20Repository/Z-Bohm/Prereq_Class/Emily%20Dashboard/PFT_Dashboard.html' ")])])
html.Div(id = 'output')])

]) #app.layout end

@app.callback(
#collect user input and assign outputs
Output(component_id = 'avg_graph', component_property = 'figure'), #This will be the dcc graph id
Input(component_id = 'slct', component_property = 'value'), #this is the input from the dropdown
)

def avg_graph(selection): #The obejct in here is what the user selects
#The function that will be below this is automatically tied to the callback and its inputs/outputs
    fig = ''
    enlisted_avg = enlisted_df.groupby('RANK')['PFT SCORE'].mean().reset_index()
    enlisted_avg['TYPE'] = 'Enlisted'
    officer_avg = officer_df.groupby('RANK')['PFT SCORE'].mean().reset_index()
    officer_avg['TYPE'] = 'Officer'
    combined_avg = pd.concat([enlisted_avg, officer_avg])
    
    if selection == '1': 
        fig = px.line(enlisted_avg, x = 'RANK', y ='PFT SCORE',
                            title = 'Average Enlisted PFT Scores', 
                            hover_name= 'PFT SCORE', 
                            markers = True, 
                            range_y = [230, 280], 
                            template = 'plotly_dark')
    elif selection == '2':
        fig = px.line(officer_avg, x = 'RANK', y ='PFT SCORE',
                            title = 'Average Officer PFT Scores', 
                            hover_name= 'PFT SCORE', 
                            markers = True, 
                            range_y = [230, 280], 
                            template = 'plotly_dark')
        
    else: 
        fig = px.line(combined_avg, x = 'RANK', y ='PFT SCORE',
                            title = 'Average Officer and Enlisted PFT Scores', 
                            hover_name= 'PFT SCORE', 
                            markers = True, 
                            range_y = [230, 280], 
                            template = 'plotly_dark',
                            line_group = 'TYPE',
                            color = 'TYPE')

    
    return fig

#############################
@app.callback(
#collect user input and assign outputs
Output(component_id = 'avg_graph2', component_property = 'figure'), #This will be the dcc graph id
Input(component_id = 'slct2', component_property = 'value'), #this is the input from the dropdown
)

def avg_graph2(selection): #The obejct in here is what the user selects
#The function that will be below this is automatically tied to the callback and its inputs/outputs
    o_avg = officer_df.groupby('PFT_DATE')['PFT SCORE'].mean().reset_index()
    o_avg.PFT_DATE = [calendar.month_name[i] for i in o_avg.PFT_DATE]

    e_avg = enlisted_df.groupby('PFT_DATE')['PFT SCORE'].mean().reset_index()
    e_avg.PFT_DATE = [calendar.month_name[i] for i in e_avg.PFT_DATE]
    
    month_avg = combined_df.groupby('PFT_DATE')['PFT SCORE'].mean().reset_index()
    month_avg.PFT_DATE = [calendar.month_name[i] for i in month_avg.PFT_DATE]

    if selection == '1': 
        fig = px.area(e_avg[:6], x = 'PFT_DATE', y = 'PFT SCORE', 
                        title = 'Average Enlisted PFT Score in Season', 
                        hover_name= 'PFT SCORE', 
                        markers = True, 
                        range_y = [235, 265], 
                        template = 'plotly_dark',
                        height = 600)
    elif selection == '2':
        fig = fig = px.area(o_avg[:6], x = 'PFT_DATE', y = 'PFT SCORE', 
                        title = 'Average Officer PFT Score in Season', 
                        hover_name= 'PFT SCORE', 
                        markers = True, 
                        range_y = [235, 275], 
                        template = 'plotly_dark',
                        height = 600)
        
    else: 
        fig = px.area(month_avg[:6], x = 'PFT_DATE', y = 'PFT SCORE', 
                        title = 'Average PFT Score in Season', 
                        hover_name= 'PFT SCORE', 
                        markers = True, 
                        range_y = [235, 265], 
                        template = 'plotly_dark',
                        height = 600)

    
    return fig
    #####################################
#Supplemental Info callback

@app.callback(
#collect user input and assign outputs
Output(component_id = 'output', component_property = 'children'), #This will be the dcc graph id
Input(component_id = 'slct3', component_property = 'value'), #this is the input from the dropdown
)

def supp(selection): #The obejct in here is what the user selects
#The function that will be below this is automatically tied to the callback and its inputs/outputs
    out = ''
    if selection == '1':
        out = ''
    if selection == '2':
        out = [html.A("Supplemental Material (Click me!)", href = 'https://dash.plotly.com/basic-callbacks', target = '_blank' )]
    else: 
        out = [html.A("Supplemental Material (Click me!)", href = 'https://dash.plotly.com/basic-callbacks', target = '_blank' )]

    
    return out






app.run_server(debug = True)
```
